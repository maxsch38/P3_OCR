## 7. <a id='toc7_'></a>[Traitement des valeurs aberrantes](#toc0_)

### 7.1. <a id='toc7_1_'></a>[Valeur energétique](#toc0_)

L'aliment le plus energétique est l'huile (100% de gras) avec 900kcal au 100g

Ici la colonne energy_100g est exprimée en kJ pour 100g

--> lien entre kcal et kJ : 1 kcal = 4,184 kJ 

Toutes la valeurs au dessus de 900*4,184 = 3765,6 kJ sont donc des valeurs aberrantes

```python
# Valeurs aberrantes : 
print(f"Il y a donc {len(df_selection.loc[df_selection['energy_100g'] > 3765.6])} produits avec une valeur energy aberrante")
```

```python
# Suppression de ces valeurs (mise à np.nan) : 
df_selection.loc[df_selection['energy_100g'] > 3765.6, 'energy_100g'] = np.nan
```

### 7.2. <a id='toc7_2_'></a>[Autres valeurs nutritionnelles](#toc0_)

```python
# Visualisation : 
liste_nutritionnelle = liste_sante[1:-3]
print("Liste des autres valeurs nutritionnelles :\n")
x =''
for i in liste_nutritionnelle: 
    x += i +' | '
print(x)
```

Ce sont des quantités pour 100g de produits, ces valeurs ne peuvent donc pas dépasser 100. 

```python
# Comptabilisation des valeurs au 100g supérieur à 100 ni y être inférieur : 
for item in liste_nutritionnelle: 
    if any(df_selection[item] > 100): 
        nb = len(df_selection.loc[df_selection[item] > 100])
        print(f"il y a {nb} valeurs aberrantes pour {item}")
```

```python
# Comptabilisation des valeurs négatives : 
for item in liste_nutritionnelle: 
    if any(df_selection[item] < 0): 
        nb = len(df_selection.loc[df_selection[item] < 0])
        print(f"il y a {nb} valeurs aberrantes pour {item}")
```

```python
# Suppression des valeurs > 100 (mise à np.nan): 
for item in liste_nutritionnelle : 
    df_selection.loc[df_selection[item] > 100, item] = np.nan
```

```python
# Suppression des valeurs < 0 (mise à np.nan): 
for item in liste_nutritionnelle : 
    df_selection.loc[df_selection[item] < 0, item] = np.nan
```

#### 7.2.2. <a id='toc7_2_2_'></a>[Somme des valeurs nutritionnelles :](#toc0_)
Les valeurs étant pour 100g de produits, leur sommes ne peuvent pas dépasser 100

```python
# Normalisation de ces valeurs entre 0 et 100 : 

# Récupération des index des lignes: 
index = list(df_selection.loc[df_selection['nutrition_sum'] > 100].index)

# Normalisation de chaque valeurs proportionnellement entre 0 et 100 : 
    # Création d'un df de travail 
df = df_selection.loc[index, liste_sante[1:-3]].copy()
sum_serie = df_selection.loc[index, 'nutrition_sum'].copy()

for col in liste_sante[1:-3]: 
    df[col] = df[col]*100/sum_serie
    
df_selection.loc[index, liste_sante[1:-3]] = df

del index, df, sum_serie
```

### 8.1. <a id='toc8_1_'></a>[Huile de palme](#toc0_)
#### 8.1.1. <a id='toc8_1_1_'></a>[Complétion à partir des labels_tags](#toc0_)

```python
# Complétion des colonnes huile de palme à partir de la colonne label_tags

# Sélection des labels_tags contenant le mot palm : 
palm = ['palm-oil', 'Palm-oil', 'huile de palme', 'Huile de palme']
ls=[]
serie = df_selection['labels_tags'].str.split(',')
serie = serie.dropna()

for label, index in zip(serie.values, serie.index): 
    if any(t in str(label) for t in palm):
        ls.append(index)

# Filtrage de serie uniquement avec les tags contenant palm :
serie = serie[ls]

# Sélection des labels_tags sans les mots de palm_free : 
palm_free = ['sans', 'Sans', 'free', 'Free', 'sustainable', 'Sustainable', 'durable', 'Durable',
             'green', 'Green']

ls = []
for item, index in zip(serie.values, serie.index): 
    for value in item: 
        if any(t in value for t in palm) and not any(t in value for t in palm_free): 
            ls.append(index)
            
# Filtrage de serie uniquement avec les tags ne contenant pas palm_free :
serie = serie[ls]

# Création de df_oil : 
df_oil = df_selection.loc[ls, ['ingredients_from_palm_oil_n', 'ingredients_that_may_be_from_palm_oil_n']].isna()

# Valeurs à 1 pour l'huile de palme : 
for index in serie.index : 
    
    if df_oil.loc[index, 'ingredients_from_palm_oil_n']: 
        df_selection.loc[index, 'ingredients_from_palm_oil_n'] = 1
        
    if df_oil.loc[index, 'ingredients_that_may_be_from_palm_oil_n']: 
        df_selection.loc[index, 'ingredients_that_may_be_from_palm_oil_n'] = 1

del ls, df_oil, serie, palm, palm_free
```

```python
# Création des colonnes label_bio et label_vege à partir de labels_tags


# Création des listes items bio et végé : 
ls_items_bio = ['Organic', 'Bio','organic', 'bio']
ls_items_vege = ['Vegan', 'Vegetarian', 'Végétarien', 'Végétalien', 
                 'vegan', 'vegetarian', 'végétarien', 'végétalien']
ls_items_bio_vege = ls_items_bio + ls_items_vege
                     

ls_index=[]
serie = df_selection['labels_tags'].str.split(',')
serie= serie.dropna()

# Création liste index : 
for label, index in zip(serie.values, serie.index): 
    if any(t in str(label) for t in ls_items_bio) & any(t in str(label) for t in ls_items_vege):
        ls_index.append(index)
        
# Filtrage de serie uniquement avec les tags ne contenant pas palm_free :
serie = serie[ls_index]

# Création des serie_label_bio et serie_label_vege
serie_label_bio = pd.Series(0, df_selection.index)
serie_label_vege = pd.Series(0, df_selection.index)
        

for label, index in zip(serie.values, serie.index): 
        if any(t in str(label) for t in ls_items_bio):
            serie_label_bio[index] = 1
        
        if any(t in str(label) for t in ls_items_vege):
            serie_label_vege[index] = 1
            
# Création des colonnes label_bio et label_vege dans df_selection : 
df_selection['label_bio'] = serie_label_bio
df_selection['label_vege'] = serie_label_vege
```

```python
# Ajout des label_bio et label_vege à la liste environnement : 
liste_environnement.append('label_bio')
liste_environnement.append('label_vege')
```

```python
# Création de colonne packaging_plastique, packaging_metal, packaging_verre, packaging_papier : 


# Création d'une série à partir de la colonne packaging_tags : 
serie = df_selection['packaging_tags']
serie = serie.str.split(',').fillna('')

# Création des listes de sélection : 
ls_plastic = ['plastique', 'plastic','Plastique', 'Plastic']
ls_metal = ['metal', 'Metal', 'métal', 'Métal', 'métale', 'Métal', 'alu', 'Alu']
ls_verre = ['verre', 'Verre', 'glass', 'Glass']
ls_papier = ['papier', 'Papier', 'carton', 'Carton', 'paper', 'Paper']

# Création des séries pour la normalisation des packaging : 
serie_plastic = pd.Series(0, df_selection.index)
serie_metal = pd.Series(0, df_selection.index)
serie_papier = pd.Series(0, df_selection.index)
serie_verre = pd.Series(0, df_selection.index)

# Complétion des séries normalisées : 
for liste, index in zip(serie.values, serie.index): 
    for tags in liste: 
        if any(t in str(tags) for t in ls_plastic):
            serie_plastic[index] = 1
            
        if any(t in str(tags) for t in ls_metal):
            serie_metal[index] = 1
            
        if any(t in str(tags) for t in ls_verre):
            serie_verre[index] = 1
            
        if any(t in str(tags) for t in ls_papier):
            serie_papier[index] = 1

# Ajout des colonnes de packaging normalisés : 
df_selection['packaging_plastique'] = serie_plastic
df_selection['packaging_metal'] = serie_metal
df_selection['packaging_verre'] = serie_verre
df_selection['packaging_papier'] = serie_papier

# Ajout à la liste environnement : 
liste_environnement.append('packaging_plastique')
liste_environnement.append('packaging_metal')
liste_environnement.append('packaging_verre')
liste_environnement.append('packaging_papier')

# Suppression de la colonne packaging_tags : 
df_selection.drop('packaging_tags', axis=1, inplace=True)

# Suppression de la liste environnement :
liste_environnement.remove('packaging_tags')

del serie, ls_metal, ls_papier, ls_plastic, ls_verre, serie_metal, serie_papier, serie_plastic, serie_verre
```
